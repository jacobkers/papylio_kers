# Getting started
A regular analysis consists of mapping the emission channels, localizing molecules and extracting the intensity traces. The data for these steps is supposed to be collected in one directory and includes a series of movie files. To get started, first import the library and create an 'Experiment'.


```python
import papylio as pp
exp = pp.Experiment('C:\Papylio example dataset')
```

When calling `pp.Experiment()` without any arguments a dialog box opens where you can select the main folder containing the single-molecule movies. Here, we use the Papylio example dataset. Alternatively, `Experiment` can be called with a specific path: `exp = pp.Experiment(r'C:\Users\you\mydir\etc\')`.

Actions

* The output text shows the experiment path. In addition, by default, the first movie file is assumed to be the one used for alignment (see below). In the example data, it is named 'Bead slide'.

* Creating the `Experiment` object will either load an existing configuration (config) file that is present in the selected path, or it will create a default config file which can then be adapted. This config file contains settings to be used in the various pipeline sections.

 The files within an `Experiment` can be listed as follows:

```python
exp.files.print()
```

Next, a series of standard processing steps can be carried out. For all these steps, intermediate results are saved into the directory as set above. Reloading the `Experiment` gives access to all intermediate results, only requiring the analysis to be performed once.
